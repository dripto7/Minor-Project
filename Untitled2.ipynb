```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score
import warnings
warnings.filterwarnings('ignore')

# Load the dataset
try:
    df = pd.read_csv('/content/spotify.csv')
    print("Dataset loaded successfully!")
except FileNotFoundError:
    print("Please ensure the CSV file is in the correct path and try again.")
    exit()

# Data Pre-processing
print("\n=== Data Pre-processing ===")

# 1. Initial data inspection
print("\n1. Initial Data Inspection:")
print(f"Dataset shape: {df.shape}")
print("\nFirst 5 rows:")
print(df.head())
print("\nData types and null values:")
print(df.info())

# 2. Handling missing values
print("\n2. Handling Missing Values:")
print("Null values before handling:")
print(df.isnull().sum())

# Drop rows with missing values (alternative could be imputation)
df = df.dropna()
print("\nNull values after handling:")
print(df.isnull().sum())

# 3. Handling duplicates
print("\n3. Handling Duplicates:")
print(f"Duplicate rows: {df.duplicated().sum()}")
df = df.drop_duplicates()
print(f"Duplicate rows after removal: {df.duplicated().sum()}")

# 4. Feature selection - we'll focus on audio features for clustering
audio_features = [
    'danceability', 'energy', 'key', 'loudness', 'mode',
    'speechiness', 'acousticness', 'instrumentalness',
    'liveness', 'valence', 'tempo', 'duration_ms'
]

# Select only the audio features and track identifiers
df_audio = df[['track_id', 'track_name', 'playlist_genre', 'playlist_subgenre'] + audio_features]

# 5. Feature scaling
scaler = StandardScaler()
scaled_features = scaler.fit_transform(df_audio[audio_features])

# Data Analysis and Visualizations
print("\n=== Data Analysis and Visualizations ===")

# 1. Distribution of genres
plt.figure(figsize=(10, 6))
sns.countplot(data=df_audio, x='playlist_genre', order=df_audio['playlist_genre'].value_counts().index)
plt.title('Distribution of Song Genres')
plt.xticks(rotation=45)
plt.show()

# 2. Distribution of subgenres (top 20)
plt.figure(figsize=(12, 8))
top_subgenres = df_audio['playlist_subgenre'].value_counts().nlargest(20).index
sns.countplot(data=df_audio, y='playlist_subgenre', order=top_subgenres)
plt.title('Top 20 Song Subgenres')
plt.show()

# 3. Correlation matrix
plt.figure(figsize=(12, 8))
corr_matrix = df_audio[audio_features].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Audio Features Correlation Matrix')
plt.show()

# 4. Pairplot of selected features (sampled for performance)
sample_df = df_audio.sample(1000)
sns.pairplot(sample_df[['danceability', 'energy', 'valence', 'acousticness', 'playlist_genre']],
             hue='playlist_genre', palette='viridis')
plt.suptitle('Pairplot of Audio Features by Genre', y=1.02)
plt.show()

# 5. Boxplots of features by genre
plt.figure(figsize=(15, 20))
for i, feature in enumerate(audio_features, 1):
    plt.subplot(4, 3, i)
    sns.boxplot(data=df_audio, x='playlist_genre', y=feature)
    plt.title(f'{feature} by Genre')
    plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Clustering Analysis
print("\n=== Clustering Analysis ===")

# 1. Determine optimal number of clusters using Elbow Method
inertia = []
silhouette_scores = []
k_range = range(2, 11)

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_features)
    inertia.append(kmeans.inertia_)

    if k > 1:  # Silhouette score requires at least 2 clusters
        silhouette_scores.append(silhouette_score(scaled_features, kmeans.labels_))

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(k_range, inertia, marker='o')
plt.title('Elbow Method for Optimal k')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')

plt.subplot(1, 2, 2)
plt.plot(range(2, 11), silhouette_scores, marker='o')
plt.title('Silhouette Scores for Optimal k')
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette Score')
plt.tight_layout()
plt.show()

# Based on the plots, choose optimal k (let's assume k=6 for this example)
optimal_k = 6

# 2. Perform K-means clustering
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
clusters = kmeans.fit_predict(scaled_features)
df_audio['cluster'] = clusters

# 3. Visualize clusters using PCA for dimensionality reduction
pca = PCA(n_components=2)
principal_components = pca.fit_transform(scaled_features)
df_audio['pca1'] = principal_components[:, 0]
df_audio['pca2'] = principal_components[:, 1]

plt.figure(figsize=(10, 8))
sns.scatterplot(data=df_audio, x='pca1', y='pca2', hue='cluster', palette='viridis', s=50)
plt.title('Song Clusters in 2D PCA Space')
plt.show()

# 4. Compare clusters with actual genres
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.countplot(data=df_audio, x='playlist_genre', hue='cluster')
plt.title('Genre Distribution in Clusters')
plt.xticks(rotation=45)

plt.subplot(1, 2, 2)
sns.countplot(data=df_audio, x='cluster', hue='playlist_genre')
plt.title('Cluster Distribution by Genre')
plt.tight_layout()
plt.show()

# 5. Analyze cluster characteristics
cluster_means = df_audio.groupby('cluster')[audio_features].mean()
plt.figure(figsize=(12, 8))
sns.heatmap(cluster_means.T, annot=True, cmap='coolwarm')
plt.title('Audio Feature Means by Cluster')
plt.show()

# 6. Example songs from each cluster
for cluster in range(optimal_k):
    print(f"\nCluster {cluster} Example Songs:")
    cluster_songs = df_audio[df_audio['cluster'] == cluster].sample(3)
    for _, song in cluster_songs.iterrows():
        print(f"{song['track_name']} - {song['playlist_genre']} (Subgenre: {song['playlist_subgenre']})")

# Final Model and Recommendations
print("\n=== Recommendation System Foundation ===")

# Function to recommend similar songs
def recommend_songs(track_id, n_recommendations=5):
    try:
        # Get the cluster of the input song
        input_cluster = df_audio[df_audio['track_id'] == track_id]['cluster'].values[0]

        # Get other songs from the same cluster
        recommendations = df_audio[df_audio['cluster'] == input_cluster].sample(n_recommendations)

        return recommendations[['track_name', 'playlist_genre', 'playlist_subgenre']]
    except:
        return "Song not found in database. Please try another track ID."

# Example recommendation
sample_track = df_audio.sample(1)['track_id'].values[0]
print(f"\nRecommendations for track ID: {sample_track}")
print(recommend_songs(sample_track))

# Save the clustered data for future use
df_audio.to_csv('spotify_songs_clustered.csv', index=False)
print("\nClustered data saved to 'spotify_songs_clustered.csv'")
```
